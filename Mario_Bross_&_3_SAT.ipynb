## 1) Introducción

Primero y antes que nada hay que tener en claro algunos conceptos como, ¿Que es un problema $NP-Hard$?, ademas de alguna pregunta o concepto que se presente mas adelante.

image.png

La imagen describe unos conjuntos y subconjubtos de problemas los cuales son los problemas $NP$, estos se subdividen en 4 grupos principales, los grupos $P$, los $NP$, tambien tenemos $NP$ complete y por ultimo los $NP-Hard$, una de las principales diferencias entre estos grupos es la velocidad con la cual se pueden resolver dichos problemas, como para los problemas de tipo $P$ estos se puden resolver en tiempo polinomial como lo puede ser algoritmos de busqueda como $Merge Sort$, $Radix Sort$, algunos problemas de optimización como el $Cutting-Road$ entre varios algoritmos mas.

Hay que mencionar un tema en especial ya que si nos fijamos en la imagen podriamos llegar a comcluir cosas tales como que $P = NP$ lo cual podira llegar a ser cierto pero lo que es mas aceptado es $P \neq NP$ ya que si fueran iguales estos terminos significaria que podemos resolver todos los problemas mas complejos en un tiempo polinomial junto a todo el grupo de los $NP$ por lo que se cree y se acepta mas que swa $P \neq NP$.
